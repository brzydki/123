```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('dataset.xlsx')
```

# Очистка датасета и нормирование для общего варианта

```python
df
```

```python
df=df.rename(columns={df.columns[0]: 'A'})
for i in range(1,10):
    df=df.rename(columns={df.columns[i]: f'{i}'})
df
```

```python
index = df[df['A'] == 'Чукотский автономный округ'].index[0]
df=df.iloc[:index+1]
```

```python
df=df.drop(labels = [0,1,2,3,4],axis = 0)
```

```python
df
```

```python
df.info
```

```python
sns.pairplot(df)
```

```python
le = LabelEncoder()
df['Region'] = le.fit_transform(df['A'])
```

```python
df
```

```python
sns.pairplot(df)
```

```python
df=df.rename(columns={df.columns[1]:"First"})
df = df.dropna(subset=['First'])
```

```python
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.Region,df.First)
```

```python
model=LinearRegression()
X=pd.DataFrame(df.Region)
y=pd.DataFrame(df.First)
model.fit(X,y)
```

```python
model.coef_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.Region,df.First)
plt.plot(X,model.predict(X),color='red')
```

```python
model.score(X,y)
```
